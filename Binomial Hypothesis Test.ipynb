```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
```

```python
n = 26 # sample size representing each individual wolf-pup
p0 = 0.5
p1 = 0.75 
a = 0.01

# upper tail of distribution
crit_value = binom.ppf(1 - a, n, p0)  # ppf is the probability mass function

# lower tail of distribution
# crit_value2 = binom.ppf(alpha, n, p0)  # ppf is the probability mass function

print(f"Critical Value: {crit_value}")

criticalRegion = 1 - binom.cdf(crit_value, n, p0)
print(criticalRegion)

# finding the power
power = 1 - binom.cdf(crit_value, n, p1)
print(f"Power: {power}")
```

```python
# create an array of possible outcomes (0 to n)
x = np.arange(0, n + 1)

# calculate the PMF for the binomial distribution under the null hypothesis (p0)
pmf_null = binom.pmf(x, n, p0)

# calculate the PMF for the binomial distribution under the alternative hypothesis (p1)
pmf_alt = binom.pmf(x, n, p1)

# plot the PMFs for both null and alternative hypotheses
plt.bar(x, pmf_null, label=f'Null Hypothesis (p={p0})', alpha=0.5)
plt.bar(x, pmf_alt, label=f'Alternative Hypothesis (p={p1})', alpha=0.5)

# highlight the critical value
plt.axvline(x=crit_value, color='red', linestyle='--', label='Critical Value')

# labeling and legend
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution')
plt.legend()

# Show the plot
plt.show()
```
